```python
import numba
import numpy as np
import galsim
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib notebook
sns.set()
```

```python
import piff
import galsim.des
from fit_des_psf import ShpPSF
```

```python
wcs = galsim.PixelScale(0.263)
pex = galsim.des.DES_PSFEx('psfcat.psf', wcs=wcs)
pff = piff.PSF.read('piff.fits')
shp = ShpPSF('shp_psf.fit')
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))

ax = axs[0]
_psf = pex.getPSF(galsim.PositionD(x=10, y=10)).drawImage(nx=17, ny=17, scale=0.263, method='no_pixel').array
ax.imshow(np.arcsinh(_psf))
ax.grid(False)
ax.set_xlabel('column [pixels]')
ax.set_ylabel('row [pixels]')

ax = axs[1]
im = galsim.ImageD(17, 17)
_psf = pff.draw(x=10, y=10, image=im).array
ax.imshow(np.arcsinh(_psf))
ax.grid(False)
ax.set_xlabel('column [pixels]')
ax.set_ylabel('row [pixels]')
plt.tight_layout()

plt.savefig('/Users/Matt/Desktop/noise.png')
```
